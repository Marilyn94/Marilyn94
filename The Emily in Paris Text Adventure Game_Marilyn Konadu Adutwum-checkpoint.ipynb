```python
"""
Welcome to the Emily in Paris Adventure Game!
 Creator: Marilyn Konadu Adutwum
 Date: January 27, 2022

"""

"""
Introduction
 
This Adventure Game is based on the Netflix Original series, Emily in Paris and it has a twist of steering up tourism. Emily is an American from Chicago who is to move to Paris to provide an American point of view to Savoir, a French marketing firm.

Emily aims at developing herself and building a strong network as she battles through work culture tolerance, adapting to a new environment and a love triangle.

Your task is to help her figure out what to do at every step of the way.

""" 

##############################################################################
# Importing the packages to aid in game
##############################################################################
import emoji

import random

##############################################################################
# Starting the Game with Stage 1 (Nested conditional statement)
##############################################################################
    
# Defining my function for stage 1
def in_Chicago():
    
    # posing the question to the user with a preamble
    print(f"""\nEmily has ben assigned to replace her boss, who is pregnant, in Paris with Savoir, a marketing firm. She has always dreamed of going to the City of Love and here lie the opportunity.
    She is contemplating on what to pack and needs some advice. As Emily’s trusted advisor,
    What would you have her pack as essentials to Paris?
    (Choose any option, 1 to 5 from the options below:)
    1. The map of France
    2. A journal
    3. A camera
    4. A book on French culture
    5. Nothing, DO NOT GO!
    """)
    
    decision = input("> \n")
    
    # converting decision to be more user friendly
    decision = decision.casefold()
    
    # creating conditional statements for responses with as many dynamics as can be
    if decision == '1' or 'map' in decision:
        print("\nAh! I see you are the geographer, Nice choice!")
        
        area_code_normandie = 2
        
        #setting a variable to enable user move on in the game
        proceed = False
        
        # creating a loop 
        while proceed == False:
            
            # an interactive geography question
            area_code = input(prompt = f"""
            What is the area code for Normandie? Clue: it's a number.
            """)
        
            # provision for possible code breakage
            try:
            # converting area_code to an integer
                area_code = int(area_code)
                
                # assigning proceed to True if it meets the integer requirement
                proceed = True
                
                # creating a conditional statement for area_code
                if area_code == area_code_normandie:
                    print("\nYou definitely know your away around France!")
            
                else: 
                    print("I am afraid that is incorrect. The correct answer is 02.")
            
                    input(prompt = ">Press enter to continue.>\n")
        
            except:
                print("Wrong entry. Kindly input a number.")
            
        # transitioning to the next stage of the adventure
        in_Paris() # This moves us into the second stage
      
    elif decision == '2' or 'journal' in decision:
        print(emoji.emojize("\nSurely, a journal can be a girl's favorite companion :winking_face:"))# get an emoji here
              
    # transitioning to the next stage of the adventure
        in_Paris() 
        
    elif decision == '3' or'camera' in decision:
        print("\nPictures indeed tell the stories!!!") 
        
    # transitioning to the next stage of the adventure
        in_Paris() # Entry to the second stage
        
    elif decision == '4' or 'book' in decision or 'culture' in decision:
        print(emoji.emojize("\nI like your style :winking_face:"))
        
    # transitioning to the next stage of the adventure
        in_Paris() # Entry to the second stage

    # defining the first fail condition
    elif decision == '5' or'Nothing' in decision or 'do not go' in decision:
        fail_1() # This moves us into fail
        
    # providing for any input out of the ordinary    
    else:
        print("Wrong entry.") 
        input (prompt = f"""
        Please press enter to try again. Be guided by the options provided. 
        You can also choose any number from 1 to 5.
        """) 
        
        # redirecting the user to the beginning to choose right
        in_Chicago() 
         
##############################################################################
# Continuing with Stage 2 
##############################################################################

# defining my function for stage 2                   
def in_Paris():
    
    # posing the question to the user with a preamble
    print(f"""\nEmily arrived safely in Paris at the CDG airport and settled in her apartment 
          with the view of the Eiffel Tour, how beautiful. Now it is time to report to work at the Savoir office.\n
          """)
    
    # the actionable question for the second scene with 4 options
    decision_2 = input(prompt = f"""
    As her chief advisor, what time would you rather she arrived at work?
    1. On time
    2. Before the work time
    3. After the work time
    4. Do not go! 
    """)
    
    # converting decision to be more user friendly
    decision_2 = decision_2.casefold()
    
    # conditionalities based on desicion_2
    if decision_2 == '1' or 'On time' in decision_2 or 'on' in decision_2:
        print("\nThat is good practice but Savoir employees may be put off by this.")
        
        # transitioning to the next stage of the game
        at_Savoir() 
        
    elif decision_2 == '2' or 'before' in decision_2 or 'early' in decision_2:
        print("\nWoooow. Relax, that will be extreme on her part for Savoir.")
        
        # transitioning to the next stage of the game
        at_Savoir()
        
    elif decision_2 == '3' or'late' in decision_2 or 'after' in decision_2:
        print(emoji.emojize("\nGreat choice :thumbs_up:"))
        
        # transitioning to the next stage of the game
        at_Savoir()
    
    elif decision_2 == '4' or 'Do not go!'in decision_2:
        print("\nYour advice got Emily in deep trouble!")
              
        # incorporating the fail() function
        fail_2()
        
    # provision for wrong input        
    else:
        print(f"""
        Your input, seems to be an error. Try again by choosing any option from the available options.
        """)
    
        # repeat of stage 2         
        in_Paris() # Entry to the third scene
              
##############################################################################
# Continuing with Stage 3 (random)
##############################################################################

# defining my function for Stage 3          
def at_Savoir():
    print(f"""\nTo blend in, Emily tries to greet her French colleagues with some gifts. 
    She only knows how to say, “Bonjour!”. Do you advise she sends gifts to her colleagues?\n
    """)
    decision_3 = input(prompt ='<Yes or No>\n')
    
    decision_3 = decision_3.casefold()
    
    # a sample of colleague gifts in a list
    gift_lst = ['croissants', 'pain au chocolat', 'flowers', 'perfume', 'a pen', 
                'pie', 'creme brulee', 'cheese', 'fish', 'macarons']
              
    # setting the lists to appear at random as suggestions to Emily
    gift_item = random.choice(gift_lst)
    
    # creating a conditional function for the response to giving gifts
    if decision_3 == '1' or 'y' in decision_3:
        
        # developing a dynamic response
        print(f"""
        Dear Emily, I suggest you gift your colleagues with {gift_item}. 
        I hear you can get some on sale, except for wine, which you will need 
        to get from Bordeaux! They will love you for sure.
        Signed,
        Your Advisor!
        """)
              
        with_friends() # Entry to the final scene
              
    # response to alternative response
    elif decision_3 == '0' or 'n' in decision_3:
        print("\nThis may create a rough path for Emily.\n")
              
        # introducing the fail function for the this stage
        fail_3()
        
              
    # bug alert
    else:
        print("\nKindly choose Yes or No to continue.\n")
              
        # repeat this stage
        at_Savoir()

##############################################################################
# Introducing the Final stage
##############################################################################
              
def with_friends():
    print(f"""\nDuring her stay, she makes two(2) friends, Mindy and Gabriel but is attracted to Gabriel who 
    has a girlfriend, Camille. She desperately needs your advice.\n
    """)
    input(prompt = 'What would you have her do? Give Emily a brief advice.')

    # finishing the game with a success
    win() 
    
##############################################################
# Creating individual fail functions for each stage
#############################################################
def fail_1():
    print(f"""I am sorry, the road comes to an end for you here. 
          Emily took your advice and no longer needs your service since she will not be travelling.
          """)
    quit()
    
def fail_2():
    print("She will have to find a more reliable advisor, but thanks for trying!")
    quit()
    
def fail_3():
    print("You've failed, but thanks for playing!")
    quit()
    
#############################################################
# Creating the win() function for the game
#############################################################
              
def win():
    print(f"""Congratulations! You are the best! Emily has garnered great skills and built good network in France. 
          You win!
          """)
    input(prompt = '<Press any key to exit\n')
    
in_Chicago() # Calling the first function in our map will start our game



    
```
